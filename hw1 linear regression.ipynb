## 1.The tensile strength of a paper product is related to the amount of hardwood in the pulp. Ten samples are produced in the pilot plant, and the data obtained are shown in the following table. 

```python
import pandas as pd

percent = pd.read_excel('hw1.xlsx')
percent
```

### (a) Fit a linear regression model relating strength to percent hardwood.


### <font color=#2093c3>ANS. percent hardwood = 0.115283strength</font> 

```python
from statsmodels.formula.api import ols

formula = 'PercentHardwood ~ Strength -1' #ที่ไม่ใช้ intercept เพราะเมื่อไม่มี strength of a paper ก็ไม่ควรมี percent hardwood

percent_l = ols(formula, data=percent).fit()

percent_l.params
```

### (b) Test the model in part (a) for significance of regression.

### <font color=#2093c3> ANS. เนื่องจากมีค่า R-squared: 0.953 และ Adjusted R-squared: 0.948 ค่อนข้างสูงแปลว่า model ค่อนข้างดี</font> 


```python
import numpy as np 

print("R-squared: {:.3f}".format(percent_l.rsquared))
print("Adjusted R-squared: {:.3f}".format(percent_l.rsquared_adj))
```

### (c)  Find a 95 percent confidence interval on the parameter

### <font color=#2093c3>ANS. ( 0.0959 , 0.1346 )</font> 



```python
percent_l.summary2()
```

## 2. A sample of 25 brands of cigarette has been tested for tar and nicotine contents, carbonmonoxide (mgs) emitted by cigarette smoke, and cigarette weight (g). Past studies have shown that the amount CO emitted increases with the tar and nicotine content.

```python
import pandas as pd

co = pd.read_excel('lect02-lin-reg.xlsx', sheet_name='CO-smoke')
co.head()
```

### a) Model the CO emitted by cigratte as a linear function of the other three quantities. Determine SSR, SSE, R2, adjusted R2, and show the fitted model coefficients.
### <font color=#2093c3>ANS. SSR= 4417.24 SSE= 45.68 R2= 0.990 adjusted R2= 0.988</font> 

### <font color=#2093c3>model: " CO =  0.936701 TAR + (-2.495939)NICOTINE + (-2.495939)WEIGHT "</font> 



```python

formula = 'CO ~ TAR + NICOTINE + WEIGHT -1 ' #เนื่องจากเป็นการทดลองจึงคิดว่าไม่มี CO จากภายนอกจึงไม่มี intercept

co_l = ols(formula, data=co).fit()

co_l.params
```

```python
print("Sum of Squared Regression (SSR): {:.2f}".format(co_l.ess)) # Explained sum of squares
print("Sum of Squared Errors (SSE): {:.2f}".format(co_l.ssr)) # Sum of Squares Residuals
print("R-squared: {:.3f}".format(co_l.rsquared))
print("Adjusted R-squared: {:.3f}".format(co_l.rsquared_adj))
```

### b) Which variables are significant and not significant in the model ?


### <font color=#2093c3>ANS. TAR and WEIGHT are significant. NICOTINE is not significant.</font> 



```python
co_l.summary2() 
```
